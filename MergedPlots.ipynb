```python
import numpy as np
from numpy.random import rand
from numpy import math
import matplotlib.pyplot as plt

nt = 100         #  number of temperature points
T = np.linspace(1.53, 3.28, nt);
```

```python
data4x4 = np.loadtxt('MagSus4x4.txt')

Magnetisation4x4 = data4x4[0]
Suszeptibility4x4 = data4x4[1]

data8x8 = np.loadtxt('MagSus8x8.txt')

Magnetisation8x8 = data8x8[0]
Suszeptibility8x8 = data8x8[1]
```

```python
f = plt.figure(figsize=(18, 10)); # plot the calculated values 

sp =  f.add_subplot(2, 2, 1 );
plt.scatter(T, abs(Magnetisation4x4) , s=50, marker='o', color='RoyalBlue')
plt.scatter(T, abs(Magnetisation8x8) , s=50, marker='^', color='g')
plt.xlabel("Temperature (T)", fontsize=20); 
plt.ylabel("Magnetization ", fontsize=20);   plt.axis('tight');

sp =  f.add_subplot(2, 2, 2 );
plt.scatter(T, Suszeptibility4x4, s=50, marker='o', color='RoyalBlue')
plt.scatter(T, Suszeptibility8x8, s=50, marker='^', color='g')
plt.xlabel("Temperature (T)", fontsize=20); 
plt.ylabel("Susceptibility", fontsize=20);   plt.axis('tight');
```
